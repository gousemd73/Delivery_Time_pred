```python
#Linear Regrssion model using the Neural Network architecture
NN_model = keras.models.Sequential()

# The Input Layer :
NN_model.add(keras.layers.Dense(128,input_dim = X_train.shape[1], activation='relu'))

# The Hidden Layers :
NN_model.add(keras.layers.Dense(256, kernel_initializer='normal',activation='relu'))
NN_model.add(keras.layers.Dense(256, kernel_initializer='normal',activation='relu'))
# NN_model.add(keras.layers.Dropout(0.3))
NN_model.add(keras.layers.Dense(256, kernel_initializer='normal',activation='relu'))
#NN_model.add(keras.layers.Dense(128,kernel_initializer='normal',activation='relu'))

# The Output Layer :
NN_model.add(keras.layers.Dense(1, kernel_initializer='normal',activation='linear'))

# Compile the network :
NN_model.compile(loss='mean_absolute_error', optimizer='adam', metrics=['mean_absolute_error'])
NN_model.summary()
```
